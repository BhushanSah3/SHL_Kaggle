```python
# === Cell 3: EDA and audio verification ===
import os
from pathlib import Path
import pandas as pd
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"  # keep your path
audios_dir = Path(DATA_DIR) / "audios"
csvs_dir = Path(DATA_DIR) / "csvs"

train_csv = csvs_dir / "train.csv"
test_csv  = csvs_dir / "test.csv"

train_df = pd.read_csv(train_csv)
test_df  = pd.read_csv(test_csv)

# Helper: map filename -> file path (handles filenames with/without .wav)
def locate_audio_file(basename, search_dir):
    # basename may be like 'audio_13_1' or 'audio_13'
    candidates = list(search_dir.rglob(f"{basename}*.wav"))
    if len(candidates) == 0:
        # try adding .wav explicitly
        p = search_dir / (basename + ".wav")
        return p if p.exists() else None
    # prefer exact match if present
    for c in candidates:
        if c.stem == basename:
            return c
    return candidates[0]

# Collect durations for train and test
def compute_durations(df, folder):
    durations = []
    missing = []
    for idx, row in df.iterrows():
        fname = str(row['filename']).strip()
        p = locate_audio_file(fname, folder)
        if p is None or not p.exists():
            missing.append(fname)
            durations.append(np.nan)
        else:
            try:
                # load with librosa (mono), sr=None to get native sampling rate
                y, sr = librosa.load(p, sr=None, mono=True)
                durations.append(len(y) / sr)
            except Exception as e:
                print("Error loading", p, ":", e)
                durations.append(np.nan)
    return durations, missing

print("Computing train durations...")
train_durs, train_missing = compute_durations(train_df, audios_dir / "train")
print("Computing test durations...")
test_durs, test_missing = compute_durations(test_df, audios_dir / "test")

train_df['duration'] = train_durs
test_df['duration'] = test_durs

# Save durations
train_df.to_csv(csvs_dir / "train_with_durations.csv", index=False)
test_df.to_csv(csvs_dir / "test_with_durations.csv", index=False)
print(f"\nSaved train_with_durations.csv and test_with_durations.csv to {csvs_dir}")

# Report missing files if any
print(f"\nMissing in train: {len(train_missing)} files")
if train_missing:
    print(train_missing[:10])
print(f"Missing in test: {len(test_missing)} files")
if test_missing:
    print(test_missing[:10])

# Basic stats
print("\nTrain label stats:")
print(train_df['label'].describe())

print("\nAudio duration stats (seconds) for train:")
print(train_df['duration'].describe())

# Visualizations
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(train_df['label'].dropna(), bins=20, kde=False)
plt.title("Train label distribution (grammar scores)")

plt.subplot(1,2,2)
sns.boxplot(x=train_df['label'])
plt.title("Label boxplot")
plt.show()

plt.figure(figsize=(10,4))
sns.histplot(train_df['duration'].dropna(), bins=30)
plt.title("Train audio durations (seconds)")
plt.xlabel("Duration (s)")
plt.show()

# Show a few sample waveforms and mel-spectrograms
def plot_wave_and_mel(path, ax_wave, ax_mel, title=None):
    y, sr = librosa.load(path, sr=16000)
    librosa.display.waveshow(y, sr=sr, ax=ax_wave)
    ax_wave.set_title(f"Waveform: {path.name}" + (f" | {title}" if title else ""))
    S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)
    S_db = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='mel', ax=ax_mel)
    ax_mel.set_title("Mel spectrogram")
    return img

# pick two example files: first non-missing, and a random one
example_files = []
for i, r in train_df.iterrows():
    if not pd.isna(r['duration']):
        example_files.append((r['filename'], Path(audios_dir / "train") / (r['filename'] + ".wav")))
    if len(example_files) >= 3:
        break

# Fallback: find actual paths using locate_audio_file
example_paths = []
for fn, _ in example_files:
    p = locate_audio_file(fn, audios_dir / "train")
    if p:
        example_paths.append(p)

# If less than 2, sample random valid ones
if len(example_paths) < 2:
    valid_paths = [locate_audio_file(fn, audios_dir / "train") for fn in train_df['filename'] if locate_audio_file(fn, audios_dir / "train")]
    valid_paths = [p for p in valid_paths if p is not None]
    example_paths = valid_paths[:2]

if example_paths:
    plt.figure(figsize=(12,6))
    ax1 = plt.subplot(2,2,1)
    ax2 = plt.subplot(2,2,2)
    ax3 = plt.subplot(2,2,3)
    ax4 = plt.subplot(2,2,4)
    try:
        plot_wave_and_mel(example_paths[0], ax1, ax2, title=f"label={train_df.loc[train_df['filename']==example_paths[0].stem,'label'].values}")
        if len(example_paths) > 1:
            plot_wave_and_mel(example_paths[1], ax3, ax4, title=f"label={train_df.loc[train_df['filename']==example_paths[1].stem,'label'].values}")
    except Exception as e:
        print("Plotting example failed:", e)
    plt.tight_layout()
    plt.show()
else:
    print("No example audio files found for plotting.")

# Quick check: are all durations between expected 45-60 seconds?
out_of_range = train_df[(train_df['duration'] < 40) | (train_df['duration'] > 70)]
print("\nNumber of train files outside expected duration range (40-70s):", len(out_of_range))
if len(out_of_range) > 0:
    display(out_of_range.head(10))

print("\nAll done. If everything looks good, we'll proceed to transcription caching (ASR) in the next cell.")
```

```python
# === Cell B_fail_convert: convert only failed filenames to 16kHz mono PCM16 ===
import subprocess
from pathlib import Path
from tqdm import tqdm
import soundfile as sf
import librosa

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
audios_dir = Path(DATA_DIR) / "audios"
processed_dir = Path(DATA_DIR) / "processed_audios"
train_src = audios_dir / "train"
test_src  = audios_dir / "test"
train_dst = processed_dir / "train"
test_dst  = processed_dir / "test"
csvs_dir = Path(DATA_DIR) / "csvs"

train_dst.mkdir(parents=True, exist_ok=True)
test_dst.mkdir(parents=True, exist_ok=True)

# Read failure list
fail_list_file = csvs_dir / "transcription_failures.txt"
if not fail_list_file.exists():
    raise FileNotFoundError(f"Failure list not found at {fail_list_file}. Run the inspection cell first.")

failed = [ln.strip() for ln in open(fail_list_file, encoding='utf-8') if ln.strip()]
print(f"Converting {len(failed)} failed files (only those).")

# Helper to find source path (train or test)
def find_src_path(basename):
    # try train then test
    for folder in (train_src, test_src):
        p = folder / (basename + ".wav")
        if p.exists():
            return p
        # fallback glob
        candidates = list(folder.rglob(f"{basename}*.wav"))
        if candidates:
            return candidates[0]
    return None

def ffmpeg_convert(src, dst):
    cmd = [
        "ffmpeg", "-y", "-i", str(src),
        "-ac", "1", "-ar", "16000", "-acodec", "pcm_s16le",
        str(dst)
    ]
    try:
        subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, check=True)
        return True
    except subprocess.CalledProcessError as e:
        print("ffmpeg failed for", src.name, "->", e)
        return False

converted = []
failed_to_convert = []
for fname in tqdm(failed):
    src = find_src_path(fname)
    if src is None:
        failed_to_convert.append((fname, "source_not_found"))
        continue
    # decide dst folder based on whether src is in train or test
    dst_folder = train_dst if str(src).startswith(str(train_src)) else test_dst
    dst = dst_folder / (Path(fname).stem + ".wav")
    ok = ffmpeg_convert(src, dst)
    if ok:
        converted.append((fname, str(dst)))
    else:
        failed_to_convert.append((fname, "ffmpeg_failed"))

print(f"\nConverted: {len(converted)} files. Failed conversions: {len(failed_to_convert)}")

# Verify first 5 converted
def verify(dst_folder, n=5):
    conv = sorted(list(dst_folder.glob("*.wav")))
    print(f"\nFound {len(conv)} files in {dst_folder}. Showing first {n}:")
    for p in conv[:n]:
        try:
            info = sf.info(str(p))
            y, sr = librosa.load(str(p), sr=None, mono=True)
            dur = len(y) / sr
            print(f"OK: {p.name} | sr={sr} | format={info.format} | duration={dur:.2f}s")
        except Exception as e:
            print("FAILED to load", p.name, e)

verify(train_dst, n=5)
verify(test_dst, n=5)

# Save a conversion log
conv_log = csvs_dir / "conversion_log.csv"
with open(conv_log, "w", encoding="utf-8") as f:
    f.write("filename,dst_path\n")
    for fn, dst in converted:
        f.write(f"{fn},{dst}\n")
print(f"\nConversion log saved to {conv_log}")
```

```python
# === Cell C_resumable_fail: Resumable ASR for only failed files (append to cache) ===
from transformers import pipeline
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import torch
import csv
import time

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
CSV_DIR = Path(DATA_DIR) / "csvs"
AUDIO_ROOT = Path(DATA_DIR) / "processed_audios"   # ensure this contains 'train' and 'test'
CACHE_PATH = CSV_DIR / "transcripts_cache.csv"
FAIL_LIST = CSV_DIR / "transcription_failures.txt"
FINAL_FAIL_LOG = CSV_DIR / "transcription_failures_final.txt"

MODEL_NAME = "openai/whisper-small"
CHUNK_LENGTH_S = 30
RETRY_COUNT = 2
SLEEP_BETWEEN_RETRIES = 3
FORCE_LANGUAGE = True

# Load list of failed filenames
failed = [ln.strip() for ln in open(FAIL_LIST, encoding='utf-8') if ln.strip()]
print("Failed list length:", len(failed))

# Prepare pipeline
device = 0 if torch.cuda.is_available() else -1
asr_kw = dict(model=MODEL_NAME, device=device, chunk_length_s=CHUNK_LENGTH_S, return_timestamps=False)
if FORCE_LANGUAGE:
    asr_kw['language'] = 'en'
asr = pipeline("automatic-speech-recognition", **asr_kw)
print("ASR pipeline ready. Device:", device)

# Ensure cache exists with header
if not CACHE_PATH.exists():
    with open(CACHE_PATH, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["filename", "transcript"])

# Read already-processed filenames in cache to skip
processed_set = set(pd.read_csv(CACHE_PATH)['filename'].astype(str).tolist())
print("Already in cache:", len(processed_set))

final_fail = []
with open(CACHE_PATH, "a", newline="", encoding="utf-8") as cache_f:
    writer = csv.writer(cache_f)
    for fname in tqdm(failed, desc="Re-ASR failed"):
        if fname in processed_set:
            continue
        # find processed path: first look in train then test
        p_train = AUDIO_ROOT / "train" / (fname + ".wav")
        p_test  = AUDIO_ROOT / "test" / (fname + ".wav")
        p = p_train if p_train.exists() else (p_test if p_test.exists() else None)
        if p is None or not p.exists():
            # fallback: look into original audios (less ideal)
            cand_train = list((Path(DATA_DIR)/"audios"/"train").rglob(f"{fname}*.wav"))
            cand_test  = list((Path(DATA_DIR)/"audios"/"test").rglob(f"{fname}*.wav"))
            p = (cand_train[0] if cand_train else (cand_test[0] if cand_test else None))
        if p is None:
            print("Missing file for re-transcription:", fname)
            writer.writerow([fname, ""])
            processed_set.add(fname)
            final_fail.append(fname)
            continue

        success = False
        transcript = ""
        for attempt in range(RETRY_COUNT+1):
            try:
                out = asr(str(p))
                if isinstance(out, dict) and 'text' in out:
                    transcript = out['text'].strip().lower()
                else:
                    transcript = str(out).strip().lower()
                success = True
                break
            except Exception as e:
                print(f"Retry ASR error for {fname} attempt {attempt+1}: {e}")
                time.sleep(SLEEP_BETWEEN_RETRIES)
        if not success:
            final_fail.append(fname)
            writer.writerow([fname, ""])
        else:
            writer.writerow([fname, transcript])
        processed_set.add(fname)

# save final fail list
with open(FINAL_FAIL_LOG, "w", encoding="utf-8") as f:
    for fn in final_fail:
        f.write(fn + "\n")

print("Re-transcription done. Appended to cache. Final fails:", len(final_fail))
print("Final fail log:", FINAL_FAIL_LOG)
```

```python
# === Cell 6: Text preprocessing, grammar & embedding features ===
# Run in Colab. May take several minutes (embedding model download + grammar-checking).

import os
from pathlib import Path
import pandas as pd
import numpy as np
import re
from tqdm import tqdm
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter
import math
import warnings
warnings.filterwarnings("ignore")

# Ensure nltk tokens are available
nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
train_path = csvs_dir / "train_with_transcripts.csv"
test_path  = csvs_dir / "test_with_transcripts.csv"
out_features = csvs_dir / "text_features.csv"
embeddings_out = csvs_dir / "text_embeddings.npy"  # numpy array aligned with features rows
embeddings_index_out = csvs_dir / "text_embeddings_index.csv"  # maps filename -> row idx in embeddings

assert train_path.exists(), f"{train_path} missing. Run ASR cell first."
assert test_path.exists(), f"{test_path} missing. Run ASR cell first."

# load
train_df = pd.read_csv(train_path)
test_df  = pd.read_csv(test_path)

# Combine for unified processing, keep indicator
train_df['split'] = 'train'
test_df['split']  = 'test'
combined = pd.concat([train_df, test_df], axis=0, ignore_index=True, sort=False)
combined['transcript'] = combined['transcript'].fillna("").astype(str)

# Basic cleaning function
FILLERS = ["um", "uh", "mm", "like", "you know", "okay", "ok", "right", "i mean"]
FILLERS_PATTERN = re.compile(r'\b(' + '|'.join([re.escape(w) for w in FILLERS]) + r')\b', flags=re.I)

def clean_text(text):
    t = text.strip().lower()
    t = re.sub(r'\s+', ' ', t)
    return t

def count_fillers(text):
    return len(FILLERS_PATTERN.findall(text))

def type_token_ratio(tokens):
    if len(tokens) == 0: return 0.0
    return len(set(tokens)) / len(tokens)

# POS ratios
def pos_ratios(tokens):
    if not tokens:
        return {"n_ratio":0.0, "v_ratio":0.0, "adj_ratio":0.0, "adv_ratio":0.0}
    tags = nltk.pos_tag(tokens)
    counts = Counter(tag for word, tag in tags)
    n = len(tokens)
    # tags starting with NN = noun, VB = verb, JJ = adj, RB = adv
    n_cnt = sum(count for tag, count in counts.items() if tag.startswith('NN'))
    v_cnt = sum(count for tag, count in counts.items() if tag.startswith('VB'))
    adj_cnt = sum(count for tag, count in counts.items() if tag.startswith('JJ'))
    adv_cnt = sum(count for tag, count in counts.items() if tag.startswith('RB'))
    return {
        "n_ratio": n_cnt / n,
        "v_ratio": v_cnt / n,
        "adj_ratio": adj_cnt / n,
        "adv_ratio": adv_cnt / n
    }

# Grammar checking setup (language_tool_python). Try to import and ensure Java available
use_language_tool = False
grammar_tool = None
try:
    import language_tool_python
    # language_tool_python requires Java runtime. Try to create a tool instance.
    try:
        grammar_tool = language_tool_python.LanguageTool('en-US')
        use_language_tool = True
        print("Using language_tool_python for grammar checks.")
    except Exception as e:
        print("language_tool_python import OK but failed to start LanguageTool (Java missing?), falling back to heuristic grammar metric.", e)
        use_language_tool = False
except Exception as e:
    print("language_tool_python not available, falling back to heuristic grammar metric.", e)
    use_language_tool = False

# Preallocate features list
features = []
filenames = combined['filename'].tolist()
durations = combined['duration'].fillna(0.0).astype(float).tolist()  # seconds

# We'll compute embeddings with sentence-transformers after text features
texts_for_embeddings = []

# Iterate and compute features
for i, row in tqdm(combined.iterrows(), total=len(combined), desc="Text features"):
    fname = row['filename']
    text = clean_text(row['transcript'])
    texts_for_embeddings.append(text)
    char_count = len(text)
    words = word_tokenize(text) if text.strip() else []
    word_count = len(words)
    sent_count = len(sent_tokenize(text)) if text.strip() else 0
    avg_sent_len = (word_count / sent_count) if sent_count>0 else 0.0
    avg_word_len = (sum(len(w) for w in words)/word_count) if word_count>0 else 0.0
    filler_count = count_fillers(text)
    ttr = type_token_ratio(words)
    pos = pos_ratios(words)
    duration_min = (row['duration'] / 60.0) if not pd.isna(row['duration']) else 0.0

    # Grammar errors
    if use_language_tool and text.strip():
        try:
            matches = grammar_tool.check(text)
            grammar_errors = len(matches)
        except Exception as e:
            grammar_errors = 0
    else:
        # fallback heuristic grammar proxy:
        # count sentences without terminal punctuation, multiple consecutive commas,
        # or many short sentences (fragments)
        grammar_errors = 0
        # heuristic 1: sentences with no end punctuation
        sents = sent_tokenize(text) if text.strip() else []
        for s in sents:
            if not re.search(r'[\.!?]$', s.strip()):
                grammar_errors += 1
        # heuristic 2: many commas (overuse)
        if text.count(',') > max(1, sent_count):
            grammar_errors += 1
        # heuristic 3: repeated filler over threshold
        if filler_count > max(1, sent_count):
            grammar_errors += 1

    grammar_errors_per_min = (grammar_errors / duration_min) if duration_min>0 else 0.0

    features.append({
        "filename": fname,
        "split": row.get('split', ''),
        "char_count": char_count,
        "word_count": word_count,
        "sent_count": sent_count,
        "avg_sent_len": avg_sent_len,
        "avg_word_len": avg_word_len,
        "filler_count": filler_count,
        "ttr": ttr,
        "n_ratio": pos['n_ratio'],
        "v_ratio": pos['v_ratio'],
        "adj_ratio": pos['adj_ratio'],
        "adv_ratio": pos['adv_ratio'],
        "grammar_errors": grammar_errors,
        "grammar_errors_per_min": grammar_errors_per_min,
        "duration_sec": row.get('duration', np.nan),
        "label": row.get('label', np.nan)
    })

# Build dataframe
feat_df = pd.DataFrame(features)
feat_df = feat_df.set_index('filename')

# Embeddings: sentence-transformers (small/fast model for reproducibility)
print("\nComputing sentence-transformer embeddings (all-MiniLM-L6-v2, 384 dims).")
from sentence_transformers import SentenceTransformer
model_name = "all-MiniLM-L6-v2"  # small and fast
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Embedding device:", device)
sbert = SentenceTransformer(model_name, device=device)

# Embed in batches to avoid OOM
BATCH = 64
all_embeddings = []
for i in tqdm(range(0, len(texts_for_embeddings), BATCH), desc="Embedding batches"):
    batch_texts = texts_for_embeddings[i:i+BATCH]
    emb = sbert.encode(batch_texts, show_progress_bar=False, convert_to_numpy=True)
    all_embeddings.append(emb)
all_embeddings = np.vstack(all_embeddings)
print("Embeddings shape:", all_embeddings.shape)

# Save embeddings and index
np.save(str(embeddings_out), all_embeddings)
pd.DataFrame({"filename": filenames, "emb_idx": list(range(len(filenames)))}).to_csv(embeddings_index_out, index=False)

# Optionally, add a few PCA components of embeddings for use in classical ML
from sklearn.decomposition import PCA
pca = PCA(n_components=16, random_state=42)
emb_pca = pca.fit_transform(all_embeddings)
for k in range(emb_pca.shape[1]):
    feat_df[f"emb_pca_{k}"] = emb_pca[:, k]

# Save features to CSV
feat_df.reset_index(inplace=True)
feat_df.to_csv(out_features, index=False)
print(f"\nSaved text features to {out_features}")
print(f"Saved embeddings to {embeddings_out} and index to {embeddings_index_out}")

# Quick preview: show train features head and some stats
display(feat_df.head(8))
print("\nText features summary:")
display(feat_df.describe().T)

# Note about grammar tool
if not use_language_tool:
    print("\nNOTE: language_tool_python was not used, falling back to heuristic grammar proxy. "
          "If you want exact grammar error counts, install Java (default-jre) in the runtime and re-run this cell:")
    print("!apt-get update && apt-get install -y default-jre -qq")
    print("Then restart runtime and re-run this cell to use language_tool_python for real grammar checks.")
```

```python
# === Cell 7: Acoustic feature extraction (MFCCs, chroma, spectral, pitch, pauses) ===
import os
from pathlib import Path
import numpy as np
import pandas as pd
import librosa
import soundfile as sf
from tqdm import tqdm
import math

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
processed_dir = Path(DATA_DIR) / "processed_audios"
audios_dir = Path(DATA_DIR) / "audios"

# Choose source folder: prefer processed (normalized) audios if present
if processed_dir.exists():
    source_dir = processed_dir
    print("Using processed_audios as source:", processed_dir)
else:
    source_dir = audios_dir
    print("Using original audios as source:", audios_dir)

# Load filenames and transcripts to compute speaking rate
transcripts_df = pd.read_csv(csvs_dir / "transcripts_cache.csv", dtype=str)
durations_df = pd.read_csv(csvs_dir / "train_with_durations.csv") if (csvs_dir / "train_with_durations.csv").exists() else None
# Merge durations for both train+test if available file train_with_durations/test_with_durations exist
combined_durations = {}
for f in [csvs_dir / "train_with_durations.csv", csvs_dir / "test_with_durations.csv"]:
    if f.exists():
        ddf = pd.read_csv(f)
        for _, r in ddf.iterrows():
            combined_durations[str(r['filename'])] = float(r.get('duration', np.nan))

# Build file list from transcripts cache (ensures alignment)
file_list = transcripts_df['filename'].tolist()

# Feature extraction params
SR = 16000
N_MFCC = 13
FRAME_LENGTH = 2048
HOP_LENGTH = 512
TOP_DB = 40  # for silence detection, lower=more aggressive silence detection

acoustic_rows = []

def load_audio(path, sr=SR):
    y, orig_sr = sf.read(path, dtype='float32')
    # if multi-channel, convert to mono
    if len(y.shape) > 1:
        y = np.mean(y, axis=1)
    if orig_sr != sr:
        y = librosa.resample(y, orig_sr, sr)
    return y, sr

for fname in tqdm(file_list, desc="Acoustic features"):
    row = {"filename": fname}
    # locate file in processed or original
    p = (source_dir / "train" / (fname + ".wav"))
    if not p.exists():
        p = (source_dir / "test" / (fname + ".wav"))
    if not p.exists():
        # fallback: glob search
        candidates = list(source_dir.rglob(f"{fname}*.wav")) + list((audios_dir / "train").rglob(f"{fname}*.wav")) + list((audios_dir / "test").rglob(f"{fname}*.wav"))
        p = candidates[0] if candidates else None
    if p is None or not p.exists():
        # missing file: fill NaNs
        row.update({
            "duration_sec": np.nan,
            "sr": np.nan,
            "mfcc_mean_0": np.nan
        })
        acoustic_rows.append(row)
        continue

    try:
        y, sr = load_audio(str(p), sr=SR)
    except Exception as e:
        # fallback to librosa load
        try:
            y, sr = librosa.load(str(p), sr=SR, mono=True)
        except Exception as e2:
            row.update({"duration_sec": np.nan})
            acoustic_rows.append(row)
            continue

    # Basic stats
    duration = len(y) / sr
    row['duration_sec'] = duration
    row['sr'] = sr
    row['rms_mean'] = float(np.mean(librosa.feature.rms(y=y, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)))
    row['rms_std'] = float(np.std(librosa.feature.rms(y=y, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)))
    row['zero_crossing_rate_mean'] = float(np.mean(librosa.feature.zero_crossing_rate(y, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)))

    # MFCCs
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC, n_fft=FRAME_LENGTH, hop_length=HOP_LENGTH)
    for i in range(N_MFCC):
        row[f"mfcc_mean_{i}"] = float(np.mean(mfcc[i]))
        row[f"mfcc_std_{i}"] = float(np.std(mfcc[i]))

    # Chroma
    try:
        chroma = librosa.feature.chroma_stft(y=y, sr=sr, n_fft=FRAME_LENGTH, hop_length=HOP_LENGTH)
        for i in range(chroma.shape[0]):
            row[f"chroma_mean_{i}"] = float(np.mean(chroma[i]))
            row[f"chroma_std_{i}"] = float(np.std(chroma[i]))
    except Exception:
        for i in range(12):
            row[f"chroma_mean_{i}"] = np.nan
            row[f"chroma_std_{i}"] = np.nan

    # Spectral contrast
    try:
        spec_contrast = librosa.feature.spectral_contrast(y=y, sr=sr, n_fft=FRAME_LENGTH, hop_length=HOP_LENGTH)
        for i in range(spec_contrast.shape[0]):
            row[f"spec_contrast_mean_{i}"] = float(np.mean(spec_contrast[i]))
            row[f"spec_contrast_std_{i}"] = float(np.std(spec_contrast[i]))
    except Exception:
        for i in range(7):
            row[f"spec_contrast_mean_{i}"] = np.nan
            row[f"spec_contrast_std_{i}"] = np.nan

    # Spectral centroid
    try:
        centroid = librosa.feature.spectral_centroid(y=y, sr=sr, n_fft=FRAME_LENGTH, hop_length=HOP_LENGTH)
        row['spec_centroid_mean'] = float(np.mean(centroid))
        row['spec_centroid_std'] = float(np.std(centroid))
    except Exception:
        row['spec_centroid_mean'] = np.nan
        row['spec_centroid_std'] = np.nan

    # Pitch estimation using librosa.pyin (may be slow); use robust try/catch
    try:
        f0, voiced_flag, voiced_prob = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'), sr=sr, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
        # f0 is array with np.nan where unvoiced
        voiced_vals = f0[~np.isnan(f0)]
        row['pitch_mean'] = float(np.mean(voiced_vals)) if len(voiced_vals) > 0 else np.nan
        row['pitch_median'] = float(np.median(voiced_vals)) if len(voiced_vals) > 0 else np.nan
        row['voiced_ratio'] = float(np.sum(~np.isnan(f0)) / len(f0))
    except Exception:
        row['pitch_mean'] = np.nan
        row['pitch_median'] = np.nan
        row['voiced_ratio'] = np.nan

    # Pause detection: non-silent intervals -> compute silence gaps
    try:
        intervals = librosa.effects.split(y, top_db=TOP_DB, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
        # intervals are (start, end) for non-silent regions in samples
        if len(intervals) <= 1:
            pause_count = 0
            total_pause = 0.0
            mean_pause = 0.0
            max_pause = 0.0
        else:
            # compute silence between non-silent segments
            pauses = []
            for i in range(1, len(intervals)):
                prev_end = intervals[i-1][1]
                cur_start = intervals[i][0]
                pause_samples = cur_start - prev_end
                pauses.append(pause_samples / sr)
            pause_count = len(pauses)
            total_pause = sum(pauses)
            mean_pause = np.mean(pauses) if pauses else 0.0
            max_pause = max(pauses) if pauses else 0.0
        row['pause_count'] = pause_count
        row['pause_total_sec'] = total_pause
        row['pause_mean_sec'] = mean_pause
        row['pause_max_sec'] = max_pause
    except Exception:
        row['pause_count'] = np.nan
        row['pause_total_sec'] = np.nan
        row['pause_mean_sec'] = np.nan
        row['pause_max_sec'] = np.nan

    # Speaking rate proxy: words per minute (use transcripts_df)
    transcript_row = transcripts_df[transcripts_df['filename'] == fname]
    if not transcript_row.empty:
        text = str(transcript_row.iloc[0]['transcript'] or "")
        word_count = 0 if text.strip() == "" else len(text.split())
        duration_min = duration / 60.0 if duration > 0 else np.nan
        words_per_min = (word_count / duration_min) if duration_min and duration_min > 0 else 0.0
        row['words_count'] = word_count
        row['words_per_min'] = words_per_min
    else:
        row['words_count'] = np.nan
        row['words_per_min'] = np.nan

    # Add basic audio file size
    try:
        row['file_size_bytes'] = p.stat().st_size
    except Exception:
        row['file_size_bytes'] = np.nan

    acoustic_rows.append(row)

# Build df and save
acoustic_df = pd.DataFrame(acoustic_rows)
acoustic_out = csvs_dir / "acoustic_features.csv"
acoustic_df.to_csv(acoustic_out, index=False)
print(f"\nSaved acoustic features to {acoustic_out}")
print("Preview:")
display(acoustic_df.head(8))
print("\nAcoustic feature extraction complete.")
```

```python
# === Cell 8: Merge text + acoustic features into final features.csv (robust) ===
import os
from pathlib import Path
import pandas as pd
import numpy as np
import json

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
models_dir = Path(DATA_DIR) / "models"
artifacts_dir = Path(DATA_DIR) / "artifacts"
models_dir.mkdir(parents=True, exist_ok=True)
artifacts_dir.mkdir(parents=True, exist_ok=True)

text_feat_path = csvs_dir / "text_features.csv"
acoustic_feat_path = csvs_dir / "acoustic_features.csv"
train_csv_path = csvs_dir / "train_with_durations.csv"
test_csv_path  = csvs_dir / "test_with_durations.csv"

# Basic existence checks
if not text_feat_path.exists():
    raise FileNotFoundError(f"{text_feat_path} not found")
if not acoustic_feat_path.exists():
    raise FileNotFoundError(f"{acoustic_feat_path} not found")

# Load
text_df = pd.read_csv(text_feat_path, dtype=str).replace({np.nan: None})
acoustic_df = pd.read_csv(acoustic_feat_path, dtype=str).replace({np.nan: None})

# Try to cast common numeric columns back to numeric, where possible
def safe_numeric_cast(df):
    for c in df.columns:
        # skip filename-like columns
        if c.lower() in ("filename", "id"):
            continue
        # attempt numeric conversion, coerce errors to NaN
        df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

text_df = safe_numeric_cast(text_df)
acoustic_df = safe_numeric_cast(acoustic_df)

# Resolve duplicate filenames in text_df by picking row with largest word_count
if "filename" not in text_df.columns:
    raise KeyError("text_features.csv must contain 'filename' column")
if "filename" not in acoustic_df.columns:
    raise KeyError("acoustic_features.csv must contain 'filename' column")

dups = text_df['filename'][text_df['filename'].duplicated()].unique().tolist()
if len(dups) > 0:
    print(f"Found {len(dups)} duplicated filenames in text_features, resolving by selecting max word_count row when available.")
    if 'word_count' in text_df.columns:
        best_rows = text_df.sort_values(['filename','word_count'], ascending=[True, False]).drop_duplicates('filename', keep='first')
    else:
        # fallback: keep first occurrence per filename
        best_rows = text_df.drop_duplicates('filename', keep='first')
    text_df = best_rows.reset_index(drop=True)
else:
    print("No duplicate filenames in text_features.")

# Ensure filename is string
text_df['filename'] = text_df['filename'].astype(str)
acoustic_df['filename'] = acoustic_df['filename'].astype(str)

# Merge on filename, outer join to keep everything
merged = pd.merge(text_df, acoustic_df, on='filename', how='outer', suffixes=('_text','_audio'))

# Attach labels if train csv exists
if train_csv_path.exists():
    train_meta = pd.read_csv(train_csv_path, dtype={'filename':str})
    if 'filename' not in train_meta.columns or 'label' not in train_meta.columns:
        print("train_with_durations.csv present but missing 'filename' or 'label' columns, skipping label attach.")
        train_meta = None
    else:
        train_meta = train_meta[['filename','label']]
        merged = merged.merge(train_meta, on='filename', how='left')
        print("Labels attached from train_with_durations.csv.")
else:
    print("train_with_durations.csv not found, labels will not be attached.")
    train_meta = None

# Add split indicator
merged['is_train'] = merged.get('label').notna() if 'label' in merged.columns else False
merged['split'] = merged['is_train'].apply(lambda x: 'train' if x else 'test')

# If 'label' exists, ensure numeric dtype, else leave absent
if 'label' in merged.columns:
    merged['label'] = pd.to_numeric(merged['label'], errors='coerce')

# Identify numeric feature columns, excluding meta columns
exclude_cols = {'filename','split','is_train','label'}
numeric_cols = []
for c in merged.columns:
    if c in exclude_cols:
        continue
    if np.issubdtype(merged[c].dtype, np.number):
        numeric_cols.append(c)

# Impute NaNs for numeric columns using median, store medians
impute_medians = {}
if len(numeric_cols) == 0:
    print("No numeric feature columns detected, nothing to impute.")
else:
    for c in numeric_cols:
        median_val = float(merged[c].median(skipna=True)) if merged[c].notna().any() else 0.0
        impute_medians[c] = median_val
        merged[c] = merged[c].fillna(median_val)

# Create a transcript_present flag if word_count exists, else infer from any text column
if 'word_count' in merged.columns:
    merged['transcript_present'] = merged['word_count'].fillna(0).astype(float) > 0
else:
    # check some common text columns
    text_like_cols = [c for c in merged.columns if any(s in c.lower() for s in ('transcript','text','asr'))]
    if len(text_like_cols) > 0:
        merged['transcript_present'] = merged[text_like_cols].notna().any(axis=1)
    else:
        merged['transcript_present'] = False

# Save final features.csv
features_out = csvs_dir / "features.csv"
merged.to_csv(features_out, index=False)
print(f"Saved merged features to {features_out}")

# Save preprocessor info for reproducibility
prep_info = {
    "numeric_cols": numeric_cols,
    "impute_medians": impute_medians,
    "features_order": merged.columns.tolist()
}
with open(models_dir / "preprocessor_info.json", "w", encoding="utf-8") as f:
    json.dump(prep_info, f, indent=2)
print(f"Saved preprocessor_info.json to {models_dir / 'preprocessor_info.json'}")

# Quick stats
print("\nMerged features shape:", merged.shape)
print("Train rows:", int(merged['is_train'].sum()), "Test rows:", int((~merged['is_train']).sum()))
# Show a quick sample
try:
    from IPython.display import display
    display(merged.head(8))
except Exception:
    print(merged.head(8))
```

```python
# === Cell 10 (force PCA on embeddings): Preprocessing pipeline for modeling ===
import os
from pathlib import Path
import json
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler, StandardScaler
from sklearn.decomposition import PCA
import joblib
import warnings
warnings.filterwarnings("ignore")

# -------- CONFIG --------
DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
models_dir = Path(DATA_DIR) / "models"
artifacts_dir = Path(DATA_DIR) / "artifacts"
models_dir.mkdir(parents=True, exist_ok=True)
artifacts_dir.mkdir(parents=True, exist_ok=True)

FEATURES_PATH = csvs_dir / "features.csv"
PREPROCESSOR_OUT = models_dir / "preprocessor.joblib"
PREPROC_META_OUT = models_dir / "preprocessor_metadata.json"
FEATURES_FOR_MODEL_CSV = csvs_dir / "features_for_model.csv"
X_TRAIN_NPY = csvs_dir / "X_train.npy"
X_TEST_NPY = csvs_dir / "X_test.npy"
Y_TRAIN_NPY = csvs_dir / "y_train.npy"

# How many PCA components to reduce embeddings to (tuneable). For 16-d -> 8 is a good start.
EMB_PCA_COMPONENTS = 8

# Scaler for numeric features
NUMERIC_SCALER = "robust"  # 'robust' or 'standard'

# -------- Load data --------
df = pd.read_csv(FEATURES_PATH)
print("Loaded features:", FEATURES_PATH, "shape:", df.shape)

# unify label column
label_candidates = [c for c in ("label","label_x","label_y") if c in df.columns]
if not label_candidates:
    raise RuntimeError("No label column found in features.csv")
label_col = label_candidates[0]
if label_col != "label":
    df["label"] = pd.to_numeric(df[label_col], errors="coerce")
else:
    df["label"] = pd.to_numeric(df["label"], errors="coerce")

# split
train_df = df[df["split"]=="train"].reset_index(drop=True)
test_df  = df[df["split"]=="test"].reset_index(drop=True)
print("Train/test sizes:", train_df.shape[0], test_df.shape[0])

# -------- Detect embedding columns (force these to be the embedding block) --------
# Heuristic: columns starting with common prefixes
embed_prefixes = ("emb_","sbert_","text_emb","embed_","pca_","sb_")
embedding_cols = [c for c in df.columns if c.lower().startswith(embed_prefixes)]
# If not found, also look for exactly 16-d numeric columns named like '0..15' (unlikely) - skip that.
print(f"Detected {len(embedding_cols)} embedding columns (will force PCA on these):", embedding_cols[:20])

# If none found, try to find columns that match pattern 'emb' anywhere
if len(embedding_cols) == 0:
    embedding_cols = [c for c in df.columns if "emb" in c.lower() or "sbert" in c.lower() or "text_emb" in c.lower()]
    print("Fallback detected embedding columns:", embedding_cols[:20])

# -------- Numeric columns: prefer preprocessor_info if available, else autodetect --------
preproc_info_path = models_dir / "preprocessor_info.json"
if preproc_info_path.exists():
    with open(preproc_info_path, "r") as f:
        prep_info = json.load(f)
    numeric_cols = prep_info.get("numeric_cols", [])
    # keep only those that exist
    numeric_cols = [c for c in numeric_cols if c in df.columns]
    print(f"Using numeric_cols from preprocessor_info.json: {len(numeric_cols)} found.")
else:
    meta_exclude = {"filename","split","is_train","label","label_x","label_y"}
    numeric_cols = [c for c in df.select_dtypes(include=[np.number]).columns.tolist() if c not in meta_exclude]
    print(f"Auto-detected numeric_cols: {len(numeric_cols)}")

# -------- Force embedding_cols to be treated as embeddings (remove them from numeric_cols if present) --------
embedding_cols = [c for c in embedding_cols if c in df.columns]
numeric_cols = [c for c in numeric_cols if c not in embedding_cols]

print(f"After forcing, numeric cols: {len(numeric_cols)}, embedding cols: {len(embedding_cols)}")

# If embedding block exists, ensure components <= embedding dim
apply_pca = len(embedding_cols) > 0 and EMB_PCA_COMPONENTS is not None
if apply_pca:
    emb_dim = len(embedding_cols)
    pca_n = min(EMB_PCA_COMPONENTS, emb_dim)
    print(f"Applying PCA to embeddings: {emb_dim} -> {pca_n} components")
else:
    pca_n = None
    if len(embedding_cols)>0:
        print("Embeddings detected but EMB_PCA_COMPONENTS is None; will passthrough embeddings.")
    else:
        print("No embeddings detected, continuing with numeric features only.")

# -------- Build transformers --------
if NUMERIC_SCALER == "robust":
    from sklearn.preprocessing import RobustScaler as _Scaler
else:
    from sklearn.preprocessing import StandardScaler as _Scaler
scaler = _Scaler()

if apply_pca:
    emb_transformer = Pipeline([("pca", PCA(n_components=pca_n, random_state=42))])
else:
    emb_transformer = "passthrough"

transformers = []
if numeric_cols:
    transformers.append(("num", scaler, numeric_cols))
if embedding_cols:
    transformers.append(("emb", emb_transformer, embedding_cols))

if len(transformers) == 0:
    raise RuntimeError("No features found to preprocess. Check numeric or embedding columns.")

from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(transformers=transformers, remainder="drop", verbose_feature_names_out=False)

# -------- Prepare feature column order (deduped) --------
feature_cols = list(dict.fromkeys(numeric_cols + embedding_cols))
print("Final feature columns count (unique):", len(feature_cols))

# ensure cols present in train/test (fill missing with zeros)
for c in feature_cols:
    if c not in train_df.columns:
        train_df[c] = 0.0
    if c not in test_df.columns:
        test_df[c] = 0.0

# -------- Fit preprocessor on train only --------
print("Fitting preprocessor on train data...")
preprocessor.fit(train_df[feature_cols])

# transform
X_train = preprocessor.transform(train_df[feature_cols])
X_test  = preprocessor.transform(test_df[feature_cols])
y_train = train_df["label"].values.astype(float)

print("Transformed shapes -> X_train:", X_train.shape, "X_test:", X_test.shape, "y_train:", y_train.shape)

# -------- Save artifacts --------
joblib.dump(preprocessor, PREPROCESSOR_OUT, compress=3)
print("Saved preprocessor to", PREPROCESSOR_OUT)

# try to get human-readable feature names
try:
    out_feature_names = preprocessor.get_feature_names_out()
    out_feature_names = [str(x) for x in out_feature_names]
except Exception:
    out_feature_names = numeric_cols.copy()
    if apply_pca:
        out_feature_names += [f"emb_pca_{i}" for i in range(pca_n)]
    else:
        out_feature_names += embedding_cols

meta = {
    "numeric_cols": numeric_cols,
    "embedding_cols": embedding_cols,
    "out_feature_names": out_feature_names,
    "applied_emb_pca": bool(apply_pca),
    "emb_pca_components": pca_n
}
with open(PREPROC_META_OUT, "w", encoding="utf-8") as f:
    json.dump(meta, f, indent=2)
print("Saved preprocessor metadata to", PREPROC_META_OUT)

np.save(X_TRAIN_NPY, X_train)
np.save(X_TEST_NPY, X_test)
np.save(Y_TRAIN_NPY, y_train)
print("Saved X_train, X_test, y_train to .npy files in csvs/")

# also attempt to save a CSV view with out_feature_names
try:
    df_train_f = pd.DataFrame(X_train, columns=out_feature_names, index=train_df.index)
    df_train_f["label"] = y_train
    df_test_f  = pd.DataFrame(X_test, columns=out_feature_names, index=test_df.index)
    df_all = pd.concat([df_train_f, df_test_f], axis=0, sort=False).reset_index(drop=True)
    df_all.to_csv(FEATURES_FOR_MODEL_CSV, index=False)
    print("Saved features_for_model.csv to", FEATURES_FOR_MODEL_CSV)
except Exception as e:
    print("Could not save human-readable features CSV:", e)

print("Preprocessing complete. Proceed to training (Cell 11).")
```

```python
# === Cell 11 (fixed RMSE calc): LightGBM combined baseline with 5-fold CV ===
import os
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold, KFold
from scipy.stats import pearsonr
import joblib
import lightgbm as lgb

# Paths
DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
models_dir = Path(DATA_DIR) / "models"
artifacts_dir = Path(DATA_DIR) / "artifacts"
models_dir.mkdir(parents=True, exist_ok=True)
artifacts_dir.mkdir(parents=True, exist_ok=True)

# Load preprocessed arrays
X_train = np.load(csvs_dir / "X_train.npy")
X_test  = np.load(csvs_dir / "X_test.npy")
y_train = np.load(csvs_dir / "y_train.npy")

# CV config
N_FOLDS = 5
RANDOM_STATE = 42

# Stratified on binned labels
try:
    y_bins = pd.qcut(y_train, q=N_FOLDS, labels=False, duplicates='drop')
    if len(np.unique(y_bins)) < N_FOLDS:
        raise Exception("qcut produced fewer bins than folds")
    skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)
    splits = skf.split(X_train, y_bins)
    print("Using StratifiedKFold on binned labels.")
except Exception:
    print("Fallback to regular KFold.")
    kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)
    splits = kf.split(X_train)

# LightGBM params
lgb_params = {
    "objective": "regression",
    "metric": "rmse",
    "learning_rate": 0.03,
    "num_leaves": 31,
    "min_data_in_leaf": 20,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.9,
    "bagging_freq": 1,
    "lambda_l1": 0.0,
    "lambda_l2": 0.1,
    "verbosity": -1,
    "seed": RANDOM_STATE,
    "n_jobs": -1
}

oof_preds = np.zeros(len(y_train))
fold_results = []

for fold, (train_idx, val_idx) in enumerate(splits):
    print(f"\n--- Fold {fold+1}/{N_FOLDS} ---")
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y_train[train_idx], y_train[val_idx]

    model = lgb.LGBMRegressor(
        n_estimators=5000,
        learning_rate=lgb_params["learning_rate"],
        num_leaves=lgb_params["num_leaves"],
        min_child_samples=lgb_params["min_data_in_leaf"],
        feature_fraction=lgb_params["feature_fraction"],
        bagging_fraction=lgb_params["bagging_fraction"],
        bagging_freq=lgb_params["bagging_freq"],
        reg_alpha=lgb_params["lambda_l1"],
        reg_lambda=lgb_params["lambda_l2"],
        random_state=RANDOM_STATE,
        n_jobs=-1
    )

    model.fit(
        X_tr, y_tr,
        eval_set=[(X_val, y_val)],
        eval_metric="rmse",
        callbacks=[lgb.early_stopping(stopping_rounds=100, verbose=True)]
    )

    val_pred = model.predict(X_val, num_iteration=model.best_iteration_)
    oof_preds[val_idx] = val_pred

    model_path = models_dir / f"lgbm_combined_fold{fold+1}.pkl"
    joblib.dump(model, model_path, compress=3)
    print(f"Saved fold model -> {model_path}")

    # robust RMSE (manual)
    rmse = float(np.sqrt(np.mean((y_val - val_pred) ** 2)))
    pearson = float(pearsonr(y_val, val_pred)[0]) if len(y_val) > 2 else float("nan")
    print(f"Fold {fold+1} RMSE: {rmse:.6f}, Pearson: {pearson:.6f}")

    fold_results.append({
        "fold": fold+1,
        "rmse": rmse,
        "pearson": pearson,
        "best_iter": int(getattr(model, "best_iteration_", -1) or -1)
    })

# Overall OOF metrics
oof_rmse = float(np.sqrt(np.mean((y_train - oof_preds) ** 2)))
try:
    oof_pearson = float(pearsonr(y_train, oof_preds)[0])
except Exception:
    oof_pearson = float("nan")
print(f"\nOOF RMSE: {oof_rmse:.6f}, OOF Pearson: {oof_pearson:.6f}")

# Save outputs
import pandas as pd, json
pd.DataFrame({"oof_pred": oof_preds, "y_true": y_train}).to_csv(csvs_dir / "oof_preds.csv", index=False)
pd.DataFrame(fold_results).to_csv(artifacts_dir / "cv_results_per_fold.csv", index=False)
with open(artifacts_dir / "cv_summary.json", "w") as f:
    json.dump({
        "oof_rmse": oof_rmse,
        "oof_pearson": oof_pearson,
        "rmse_mean": float(pd.Series([r["rmse"] for r in fold_results]).mean()),
        "rmse_std": float(pd.Series([r["rmse"] for r in fold_results]).std()),
        "pearson_mean": float(pd.Series([r["pearson"] for r in fold_results]).mean()),
        "pearson_std": float(pd.Series([r["pearson"] for r in fold_results]).std())
    }, f, indent=2)

print("Saved OOF preds and CV results.")
print("Cell 11 complete.")
```

```python
# Repair + Preprocess attempt (best-effort)
import pandas as pd, numpy as np, json, joblib
from pathlib import Path
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
models_dir = Path(DATA_DIR) / "models"

clean_path = csvs_dir / "features_for_model_clean.csv"
orig_path  = csvs_dir / "features.csv"

print("Paths:")
print(" cleaned:", clean_path.exists(), clean_path)
print(" original:", orig_path.exists(), orig_path)
print()

df_clean = pd.read_csv(clean_path)
print("Cleaned shape:", df_clean.shape)
print("Cleaned columns (first 30):", df_clean.columns.tolist()[:30])
print(df_clean.head(3))
print()

if orig_path.exists():
    df_orig = pd.read_csv(orig_path)
    print("Original shape:", df_orig.shape)
    print("Original columns (first 50):", df_orig.columns.tolist()[:50])
    print(df_orig.head(3))
else:
    df_orig = None
    print("Original features.csv not found. We will not be able to recover filename->label mapping automatically.")
print("\n---\n")

# Helper to check common keys
def common_cols(a, b):
    return [c for c in a.columns if c in b.columns]

if df_orig is not None:
    commons = common_cols(df_clean, df_orig)
    print("Common columns between cleaned and original:", commons)

# Attempt 1: if 'filename' exists in original and not in cleaned, but shapes equal -> align by index
repaired = df_clean.copy()
if df_orig is not None and 'filename' in df_orig.columns and 'label' in df_orig.columns:
    if 'filename' not in repaired.columns and repaired.shape[0] == df_orig.shape[0]:
        print("No filename in cleaned but shapes match; copying 'filename' and 'label' from original by index (best-effort).")
        repaired['filename'] = df_orig['filename'].values
        repaired['label'] = df_orig['label'].values
    elif 'filename' in repaired.columns:
        print("'filename' already present in cleaned; mapping labels from original by filename.")
        # safe merge on filename
        map_lab = df_orig.set_index('filename')['label'].to_dict()
        repaired['label'] = repaired['filename'].map(map_lab)
    else:
        print("Shapes do not match or no filename in original; cannot auto-copy filename/label by index.")
else:
    print("Original file missing or lacks filename/label columns; cannot auto-repair.")

# If still no label, try to find any other linking column like 'id' or 'utterance' etc
if 'label' not in repaired.columns or repaired['label'].isnull().all():
    print("Label still missing or all-NaN after index attempt. Searching for alternate link columns in original...")
    if df_orig is not None:
        potential_keys = ['id','file_id','utterance','utt','audio_id','name']
        found = False
        for k in potential_keys:
            if k in df_orig.columns:
                print("Found potential key in original:", k)
                # if cleaned has same key, merge on it
                if k in repaired.columns:
                    print("Merging on", k)
                    repaired = repaired.merge(df_orig[[k,'label']], on=k, how='left')
                    found = True
                    break
        if not found:
            print("No alternate key found in common. Last resort: if shapes match you can align by index manually.")
    else:
        print("No original to search for alternate keys.")

# Final checks
print("\nAfter repair attempt: columns:", repaired.columns.tolist()[:40])
if 'label' in repaired.columns:
    print("label unique:", repaired['label'].nunique(), "nulls:", repaired['label'].isnull().sum())
else:
    print("label still not present. STOP and inspect mapping manually.")

# If label exists, continue with preprocessing pipeline (same logic as your Cell A)
if 'label' in repaired.columns and repaired['label'].notna().sum() > 0:
    # mark train/test
    train_mask = repaired['label'].notna()
    train_df = repaired[train_mask].copy()
    test_df = repaired[~train_mask].copy()
    print("Train rows:", train_df.shape[0], "Test rows:", test_df.shape[0])

    # numeric columns
    numeric_cols = [c for c in repaired.columns if np.issubdtype(repaired[c].dtype, np.number) and c not in ('label',)]
    print("Numeric cols detected:", len(numeric_cols))
    X_train_base = train_df[numeric_cols].copy()
    X_test_base = test_df[numeric_cols].copy()

    # If embeddings present, detect & PCA (heuristic)
    embedding_cols = [c for c in repaired.columns if c.startswith('emb_') or 'sbert' in c.lower() or 'embedding' in c.lower()]
    print("Embedding cols found:", len(embedding_cols))

    if len(embedding_cols) > 0:
        all_emb = pd.concat([train_df[embedding_cols].fillna(0), test_df[embedding_cols].fillna(0)], axis=0).values
        pca_n = max(8, min(32, len(embedding_cols)//2))
        pca = PCA(n_components=pca_n, random_state=42)
        pca.fit(all_emb)
        train_p = pca.transform(train_df[embedding_cols].fillna(0).values)
        test_p = pca.transform(test_df[embedding_cols].fillna(0).values)
        X_train_base = X_train_base.drop(columns=embedding_cols, errors='ignore')
        X_test_base = X_test_base.drop(columns=embedding_cols, errors='ignore')
        for i in range(train_p.shape[1]):
            X_train_base[f'emb_pca_{i}'] = train_p[:, i]
            X_test_base[f'emb_pca_{i}'] = test_p[:, i]
        joblib.dump(pca, models_dir / "embeddings_pca.joblib", compress=3)

    # Preprocessing pipeline
    imputer = SimpleImputer(strategy='median')
    scaler = RobustScaler()
    pipe = Pipeline([('imputer', imputer), ('scaler', scaler)])
    pipe.fit(X_train_base)
    X_train = pipe.transform(X_train_base)
    X_test = pipe.transform(X_test_base)
    y_train = train_df['label'].values.astype(float)

    # Save artifacts
    joblib.dump(pipe, models_dir / "preprocessor_repaired.joblib", compress=3)
    meta = {
        "numeric_cols": X_train_base.columns.tolist(),
        "embedding_cols_after_pca": [c for c in X_train_base.columns if c.startswith('emb_pca_')],
        "impute_strategy": "median",
        "scaler": "RobustScaler"
    }
    json.dump(meta, open(models_dir / "preprocessor_metadata_repaired.json","w"), indent=2)

    np.save(csvs_dir / "X_train.npy", X_train)
    np.save(csvs_dir / "X_test.npy", X_test)
    np.save(csvs_dir / "y_train.npy", y_train)
    pd.DataFrame(X_train).to_csv(csvs_dir / "X_train_preview.csv", index=False)
    pd.DataFrame(X_test).to_csv(csvs_dir / "X_test_preview.csv", index=False)
    train_df.to_csv(csvs_dir / "train_rows_used_repaired.csv", index=False)

    print("Preprocessing succeeded. shapes:", X_train.shape, X_test.shape, y_train.shape)
else:
    print("Repair failed: label column missing. Please inspect the printed diagnostics above and paste them here so I can help further.")
```

```python
# Full Cell B: LightGBM CV (stable version for all sklearn/lightgbm versions)
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import lightgbm as lgb
from lightgbm import early_stopping, log_evaluation
import gc, json, os

# Paths
csvs_dir = Path("/content/drive/MyDrive/SHL_dataset/dataset/csvs")
models_dir = Path("/content/drive/MyDrive/SHL_dataset/dataset/models")
os.makedirs(models_dir, exist_ok=True)

# Load arrays
X_train = np.load(csvs_dir / "X_train.npy")
X_test  = np.load(csvs_dir / "X_test.npy")
y       = np.load(csvs_dir / "y_train.npy")

print("Loaded shapes:", X_train.shape, X_test.shape, y.shape)

# Stratify bins for regression
n_bins = 5
try:
    y_bins = pd.qcut(y, q=n_bins, labels=False, duplicates='drop')
except Exception:
    y_bins = pd.cut(y, bins=n_bins, labels=False)

n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

oof_preds = np.zeros(len(y))
feature_importances = []
fold_metrics = []

# LightGBM params
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'min_child_samples': 20,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 1,
    'verbosity': -1,
    'seed': 42,
    'n_jobs': -1
}

# CV loop
for fold, (train_idx, val_idx) in enumerate(skf.split(X_train, y_bins), start=1):
    print(f"\n===== Fold {fold} =====")
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y[train_idx], y[val_idx]

    lgb_train = lgb.Dataset(X_tr, label=y_tr)
    lgb_val   = lgb.Dataset(X_val, label=y_val, reference=lgb_train)

    model = lgb.train(
        params,
        lgb_train,
        num_boost_round=5000,
        valid_sets=[lgb_train, lgb_val],
        valid_names=['train','valid'],
        callbacks=[
            early_stopping(stopping_rounds=100, verbose=True),
            log_evaluation(period=100)
        ]
    )

    # Predictions and metrics
    val_pred = model.predict(X_val, num_iteration=model.best_iteration)
    oof_preds[val_idx] = val_pred

    # Compatible RMSE computation
    rmse = np.sqrt(mean_squared_error(y_val, val_pred))
    pearson = pearsonr(y_val, val_pred)[0]
    print(f"Fold {fold} | RMSE: {rmse:.4f} | Pearson: {pearson:.4f} | best_iter: {model.best_iteration}")
    fold_metrics.append((fold, float(rmse), float(pearson), int(model.best_iteration)))

    # Feature importance
    try:
        fi = model.feature_importance(importance_type='gain')
    except Exception:
        fi = model.feature_importance()
    feature_importances.append(fi)

    model.save_model(str(models_dir / f"lgb_fold{fold}.txt"))
    del model, lgb_train, lgb_val
    gc.collect()

# Overall OOF metrics
oof_rmse = np.sqrt(mean_squared_error(y, oof_preds))
oof_pearson = pearsonr(y, oof_preds)[0]

print("\nPer-fold metrics (fold, rmse, pearson, best_iter):")
for fm in fold_metrics:
    print(fm)
print(f"\nOOF RMSE: {oof_rmse:.4f}")
print(f"OOF Pearson: {oof_pearson:.4f}")

# Save OOF preds
np.save(csvs_dir / "oof_preds.npy", oof_preds)
pd.DataFrame({"y_true": y, "y_pred": oof_preds}).to_csv(csvs_dir / "oof_preds.csv", index=False)

# Aggregate feature importance and save
meta_paths = [
    models_dir / "preprocessor_metadata.json",
    models_dir / "preprocessor_metadata_repaired.json",
    models_dir / "preprocessor_info.json"
]
feat_names = None
for mp in meta_paths:
    if mp.exists():
        try:
            meta = json.load(open(mp))
            feat_names = meta.get("numeric_cols", None)
            if feat_names:
                break
        except Exception:
            feat_names = None

if feat_names is None:
    feat_names = [f"f{i}" for i in range(X_train.shape[1])]

fi_mean = np.mean(np.vstack(feature_importances), axis=0)
fi_df = pd.DataFrame({"feature": feat_names, "gain": fi_mean})
fi_df = fi_df.sort_values("gain", ascending=False).reset_index(drop=True)
fi_df.to_csv(csvs_dir / "feature_importance_cv.csv", index=False)

print("\nSaved OOF preds and feature importance to:", csvs_dir)
```

```python
# STEP 1: DROP leftover label-like columns and re-run preprocessing (single cell)
import os, json
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
import joblib

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
models_dir = Path(DATA_DIR) / "models"
artifacts_dir = Path(DATA_DIR) / "artifacts"
os.makedirs(models_dir, exist_ok=True)
os.makedirs(artifacts_dir, exist_ok=True)

# file paths
feat_clean_path = csvs_dir / "features_for_model_clean.csv"
feat_v2_path    = csvs_dir / "features_for_model_clean_v2.csv"
orig_feat_path  = csvs_dir / "features.csv"

print("Paths exist? cleaned:", feat_clean_path.exists(), " original:", orig_feat_path.exists())

# 1) Load cleaned features
df = pd.read_csv(feat_clean_path)
print("Loaded cleaned features shape:", df.shape)
print("Columns sample:", df.columns.tolist()[:60])

# 2) Drop any label-like or merge-artifact columns
drop_patterns = ['label', 'target', 'y_true', 'y_pred']
# also drop typical pandas merge suffixes if present
drop_cols = [c for c in df.columns if any(p in c.lower() for p in drop_patterns) or c.lower().endswith('_x') or c.lower().endswith('_y')]
# but be careful: keep numeric features only; we will not drop columns like 'pca0' etc
# remove false-positives like column names that end with '_x' but are genuine features (rare). We assume none.
print("Dropping columns (detected):", drop_cols)
df = df.drop(columns=drop_cols, errors='ignore')

# Save new cleaned CSV
df.to_csv(feat_v2_path, index=False)
print("Saved cleaned v2 to:", feat_v2_path, " new shape:", df.shape)

# 3) Reattach label mapping from original features.csv if necessary
orig = None
if orig_feat_path.exists():
    orig = pd.read_csv(orig_feat_path)
    print("Loaded original features.csv shape:", orig.shape)
else:
    print("Original features.csv not found:", orig_feat_path)

# If df has no 'label' or no 'filename', attempt safe mapping by index if shapes match
repaired = df.copy()
if 'label' not in repaired.columns:
    if orig is not None and 'label' in orig.columns:
        # prefer mapping via filename if cleaned has filename
        if 'filename' in repaired.columns and 'filename' in orig.columns:
            map_lab = orig.set_index('filename')['label'].to_dict()
            repaired['label'] = repaired['filename'].map(map_lab)
            print("Mapped labels from original via filename. Missing labels:", repaired['label'].isna().sum())
        else:
            # last-resort index-based copy only if shapes match exactly
            if repaired.shape[0] == orig.shape[0]:
                repaired['filename'] = orig['filename'].values
                repaired['label'] = orig['label'].values
                print("Copied filename and label by index from original (shapes match).")
            else:
                raise RuntimeError("Cannot reattach labels: no filename in cleaned and shapes mismatch with original. Inspect files manually.")

# Final check
if 'label' not in repaired.columns:
    raise RuntimeError("Label column still missing after attempts. Stop and inspect.")
print("After attach: label unique:", repaired['label'].nunique(), "nulls:", repaired['label'].isna().sum())

# 4) Split into train/test rows
train_mask = repaired['label'].notna()
train_df = repaired[train_mask].copy()
test_df  = repaired[~train_mask].copy()
print("Train rows:", train_df.shape[0], "Test rows:", test_df.shape[0])

# 5) Numeric columns to use (exclude label)
excluded = {'label','filename','split','is_train'}
numeric_cols = [c for c in repaired.columns if c not in excluded and pd.api.types.is_numeric_dtype(repaired[c])]
print("Numeric columns count:", len(numeric_cols))

X_train_base = train_df[numeric_cols].copy()
X_test_base  = test_df[numeric_cols].copy()

# 6) Detect embedding columns heuristically and apply PCA if found (keeps runtime manageable)
embedding_cols = [c for c in repaired.columns if c.startswith('emb_') or 'sbert' in c.lower() or 'embedding' in c.lower()]
print("Detected embedding columns:", len(embedding_cols))

if len(embedding_cols) > 0:
    all_emb = pd.concat([train_df[embedding_cols].fillna(0), test_df[embedding_cols].fillna(0)], axis=0).values
    pca_n = max(8, min(32, len(embedding_cols)//2))
    pca = PCA(n_components=pca_n, random_state=42)
    pca.fit(all_emb)
    train_p = pca.transform(train_df[embedding_cols].fillna(0).values)
    test_p  = pca.transform(test_df[embedding_cols].fillna(0).values)
    X_train_base = X_train_base.drop(columns=embedding_cols, errors='ignore')
    X_test_base  = X_test_base.drop(columns=embedding_cols, errors='ignore')
    for i in range(train_p.shape[1]):
        X_train_base[f'emb_pca_{i}'] = train_p[:, i]
        X_test_base[f'emb_pca_{i}']  = test_p[:, i]
    joblib.dump(pca, models_dir / "embeddings_pca.joblib", compress=3)
    print("Applied PCA to embeddings, components:", train_p.shape[1])

# 7) Fit preprocessing pipeline (median impute + robust scaler)
imputer = SimpleImputer(strategy='median')
scaler  = RobustScaler()
pipe = Pipeline([('imputer', imputer), ('scaler', scaler)])
pipe.fit(X_train_base)
X_train = pipe.transform(X_train_base)
X_test  = pipe.transform(X_test_base)
y_train = train_df['label'].values.astype(float)

# Save artifacts
joblib.dump(pipe, models_dir / "preprocessor_v2.joblib", compress=3)
meta = {
    "numeric_cols": X_train_base.columns.tolist(),
    "embedding_cols_after_pca": [c for c in X_train_base.columns if c.startswith('emb_pca_')],
    "impute_strategy": "median",
    "scaler": "RobustScaler"
}
json.dump(meta, open(models_dir / "preprocessor_metadata_v2.json","w"), indent=2)

np.save(csvs_dir / "X_train.npy", X_train)
np.save(csvs_dir / "X_test.npy", X_test)
np.save(csvs_dir / "y_train.npy", y_train)

pd.DataFrame(X_train).to_csv(csvs_dir / "X_train_preview.csv", index=False)
pd.DataFrame(X_test).to_csv(csvs_dir / "X_test_preview.csv", index=False)
train_df.to_csv(csvs_dir / "train_rows_used_repaired_v2.csv", index=False)

print("Preprocessing v2 done. Shapes:", X_train.shape, X_test.shape, y_train.shape)
print("Saved files: X_train.npy, X_test.npy, y_train.npy, preprocessor_v2, preprocessor_metadata_v2.json")
```

```python
# Experiment 3A: compute LM perplexity + GEC edit counts and append to features CSV
# Run as a single cell. Requires: transformers, torch, tqdm
import os, sys, math
from pathlib import Path
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Paths
DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
feat_in  = csvs_dir / "features_for_model_clean_v2.csv"
feat_out = csvs_dir / "features_for_model_enriched_text.csv"

if not feat_in.exists():
    raise FileNotFoundError(f"Input features CSV not found: {feat_in}")

# Load dataframe
df = pd.read_csv(feat_in)
print("Loaded:", feat_in, "shape:", df.shape)

# 1) detect transcript column
possible_text_cols = ['transcript', 'text', 'transcript_text', 'utterance', 'transcript_clean']
text_col = None
for c in possible_text_cols:
    if c in df.columns:
        text_col = c
        break
# sometimes transcripts are not present as complete text; try 'words' or 'words_str' or 'transcript_joined'
fallbacks = ['words', 'words_str', 'transcript_joined']
if text_col is None:
    for c in df.columns:
        if 'transcript' in c.lower() or 'text' in c.lower() or 'utter' in c.lower():
            text_col = c
            break

if text_col is None:
    print("No transcript column auto-detected. Please ensure your CSV has a transcript column.")
    print("Columns:", df.columns.tolist())
    raise RuntimeError("Transcript column not found.")

print("Using transcript column:", text_col)

# Device selection
device = 0 if torch.cuda.is_available() else -1
print("Using device:", "cuda" if device==0 else "cpu")

# 2) Load LM (causal) for perplexity. Use distilgpt2 for speed
lm_name = "distilgpt2"
print("Loading LM:", lm_name)
try:
    lm_tokenizer = AutoTokenizer.from_pretrained(lm_name)
    lm_model = AutoModelForCausalLM.from_pretrained(lm_name).to("cuda" if device==0 else "cpu")
    lm_model.eval()
except Exception as e:
    print("Failed to load LM model:", e)
    raise

# helper: compute loss for a single string (auto truncation/padding)
def lm_loss_for_text(text, tokenizer, model, max_length=512):
    # prepare inputs
    enc = tokenizer(text, return_tensors="pt", truncation=True, max_length=max_length)
    input_ids = enc["input_ids"].to(model.device)
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        # outputs.loss is mean token loss
        loss = outputs.loss.item()
    return loss

# 3) Load GEC model via text2text pipeline
# Preferred: prithivida/grammar_error_correcter_v1 (small), else try "pszemraj/grammar-error-correction"
gec_models = ["prithivida/grammar_error_correcter_v1", "pszemraj/grammar-error-correction", "facebook/bart-large"]
gec_model_name = None
for gm in gec_models:
    try:
        gec_pipe = pipeline("text2text-generation", model=gm, device=0 if device==0 else -1, truncation=True)
        gec_model_name = gm
        print("Loaded GEC model:", gm)
        break
    except Exception as e:
        print(f"Could not load GEC model {gm}: {e}")
        gec_pipe = None
        continue

if gec_model_name is None:
    print("No GEC model available. Skipping GEC step.")
    gec_pipe = None

# 4) Prepare output columns
lm_losses = []
lm_perps = []
gec_corrected = []
gec_edits = []
gec_edits_per_min = []

# get duration for per-minute normalization
duration_col = None
for c in ['duration_sec_text', 'duration_sec_audio', 'duration_sec']:
    if c in df.columns:
        duration_col = c
        break

# helper: word-level edit count using difflib
import difflib
def word_edit_count(a, b):
    # token-level (words)
    a_tokens = str(a).strip().split()
    b_tokens = str(b).strip().split()
    # use SequenceMatcher to compute opcodes
    sm = difflib.SequenceMatcher(a=a_tokens, b=b_tokens)
    edits = 0
    for tag, i1, i2, j1, j2 in sm.get_opcodes():
        if tag in ("replace", "delete", "insert"):
            # count number of token changes
            if tag == "replace":
                edits += max(i2 - i1, j2 - j1)
            elif tag == "delete":
                edits += (i2 - i1)
            elif tag == "insert":
                edits += (j2 - j1)
    return edits

# iterate rows (use tqdm)
rows = df.shape[0]
for idx, row in tqdm(df.iterrows(), total=rows):
    text = str(row[text_col]) if pd.notna(row[text_col]) else ""
    if len(text.strip()) == 0:
        lm_losses.append(np.nan)
        lm_perps.append(np.nan)
        gec_corrected.append("")
        gec_edits.append(np.nan)
        gec_edits_per_min.append(np.nan)
        continue

    # LM loss/perplexity (try/except to be robust)
    try:
        loss = lm_loss_for_text(text, lm_tokenizer, lm_model, max_length=512)
        perp = math.exp(loss) if loss < 100 else float("inf")
    except Exception as e:
        print("LM error at idx", idx, e)
        loss = np.nan
        perp = np.nan

    lm_losses.append(loss)
    lm_perps.append(perp)

    # GEC correction and edit count
    if gec_pipe is None:
        gec_corrected.append("")
        gec_edits.append(np.nan)
        gec_edits_per_min.append(np.nan)
    else:
        try:
            # generate correction (may be slow)
            out = gec_pipe(text, max_length=512, truncation=True)
            corrected = out[0]["generated_text"] if isinstance(out, list) else str(out)
            gec_corrected.append(corrected)
            edits = word_edit_count(text, corrected)
            gec_edits.append(edits)
            # normalize per minute
            if duration_col is not None and pd.notna(row[duration_col]) and row[duration_col] > 0:
                per_min = edits / (row[duration_col] / 60.0)
            else:
                per_min = np.nan
            gec_edits_per_min.append(per_min)
        except Exception as e:
            print("GEC error at idx", idx, e)
            gec_corrected.append("")
            gec_edits.append(np.nan)
            gec_edits_per_min.append(np.nan)

# attach to dataframe
df['lm_loss'] = lm_losses
df['lm_perplexity'] = lm_perps
df['gec_corrected'] = gec_corrected
df['gec_edits'] = gec_edits
df['gec_edits_per_min'] = gec_edits_per_min

# Save enriched CSV
df.to_csv(feat_out, index=False)
print("Saved enriched features to:", feat_out)
print("Preview of new columns:")
print(df[['lm_loss','lm_perplexity','gec_edits','gec_edits_per_min']].head(10))
```

```python
# STEP 2: Preprocess enriched CSV (build X_train/X_test/y_train including LM/GEC features)
import os, json
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
import joblib

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
models_dir = Path(DATA_DIR) / "models"
os.makedirs(models_dir, exist_ok=True)

feat_enriched = csvs_dir / "features_for_model_enriched_text.csv"
if not feat_enriched.exists():
    raise FileNotFoundError(f"Enriched features CSV not found: {feat_enriched}")

df = pd.read_csv(feat_enriched)
print("Loaded enriched CSV:", feat_enriched, "shape:", df.shape)
print("Sample columns:", df.columns.tolist()[:80])

# 1) Drop any remaining label-like or merge-artifact columns explicitly
drop_patterns = ['label', 'target', 'y_true', 'y_pred']
drop_cols = [c for c in df.columns if any(p in c.lower() for p in drop_patterns) or c.lower().endswith('_x') or c.lower().endswith('_y')]
# Ensure we do NOT drop legitimate features like 'lm_loss' etc (they don't match patterns)
print("Dropping these columns if present:", drop_cols)
df = df.drop(columns=drop_cols, errors='ignore')

# 2) Ensure label is present by mapping from original if needed
orig_path = csvs_dir / "features.csv"
if 'label' not in df.columns:
    if orig_path.exists():
        orig = pd.read_csv(orig_path)
        if 'filename' in df.columns and 'filename' in orig.columns:
            df['label'] = df['filename'].map(orig.set_index('filename')['label'])
            print("Mapped label via filename. Missing labels:", df['label'].isna().sum())
        elif df.shape[0] == orig.shape[0]:
            df['label'] = orig['label'].values
            df['filename'] = orig['filename'].values
            print("Copied filename and label by index from original.")
        else:
            raise RuntimeError("Label missing and cannot be mapped automatically. Inspect files.")
    else:
        raise RuntimeError("Label missing and original features.csv not found.")

# 3) Split train/test
train_mask = df['label'].notna()
train_df = df[train_mask].copy()
test_df  = df[~train_mask].copy()
print("Train rows:", train_df.shape[0], "Test rows:", test_df.shape[0])

# 4) Identify numeric columns (keep the new LM/GEC numeric cols)
excluded = {'label','filename','split','is_train','gec_corrected'}  # geop_corrected is text, exclude
numeric_cols = [c for c in df.columns if c not in excluded and pd.api.types.is_numeric_dtype(df[c])]
print("Detected numeric columns count:", len(numeric_cols))
print("Some numeric columns sample:", numeric_cols[:50])

# 5) Prepare X bases
X_train_base = train_df[numeric_cols].copy()
X_test_base  = test_df[numeric_cols].copy()

# 6) Embedding PCA if any embeddings present
embedding_cols = [c for c in df.columns if c.startswith('emb_') or 'sbert' in c.lower() or 'embedding' in c.lower()]
print("Detected embedding columns:", embedding_cols)
if len(embedding_cols) > 0:
    all_emb = pd.concat([train_df[embedding_cols].fillna(0), test_df[embedding_cols].fillna(0)], axis=0).values
    pca_n = max(8, min(32, len(embedding_cols)//2))
    pca = PCA(n_components=pca_n, random_state=42)
    pca.fit(all_emb)
    train_p = pca.transform(train_df[embedding_cols].fillna(0).values)
    test_p  = pca.transform(test_df[embedding_cols].fillna(0).values)
    X_train_base = X_train_base.drop(columns=embedding_cols, errors='ignore')
    X_test_base  = X_test_base.drop(columns=embedding_cols, errors='ignore')
    for i in range(train_p.shape[1]):
        X_train_base[f'emb_pca_{i}'] = train_p[:, i]
        X_test_base[f'emb_pca_{i}'] = test_p[:, i]
    joblib.dump(pca, models_dir / "embeddings_pca_enriched.joblib", compress=3)
    print("Applied PCA to embeddings, components:", train_p.shape[1])

# 7) Fit imputer+scaler pipeline on train
imputer = SimpleImputer(strategy='median')
scaler  = RobustScaler()
pipe = Pipeline([('imputer', imputer), ('scaler', scaler)])
pipe.fit(X_train_base)
X_train = pipe.transform(X_train_base)
X_test  = pipe.transform(X_test_base)
y_train = train_df['label'].values.astype(float)

# 8) Save artifacts
joblib.dump(pipe, models_dir / "preprocessor_enriched.joblib", compress=3)
meta = {
    "numeric_cols": X_train_base.columns.tolist(),
    "embedding_cols_after_pca": [c for c in X_train_base.columns if c.startswith('emb_pca_')],
    "impute_strategy": "median",
    "scaler": "RobustScaler"
}
json.dump(meta, open(models_dir / "preprocessor_metadata_enriched.json","w"), indent=2)

np.save(csvs_dir / "X_train.npy", X_train)
np.save(csvs_dir / "X_test.npy", X_test)
np.save(csvs_dir / "y_train.npy", y_train)

# previews and saved train rows used
pd.DataFrame(X_train).to_csv(csvs_dir / "X_train_preview_enriched.csv", index=False)
pd.DataFrame(X_test).to_csv(csvs_dir / "X_test_preview_enriched.csv", index=False)
train_df.to_csv(csvs_dir / "train_rows_used_enriched.csv", index=False)

print("Preprocessing (enriched) done. Shapes:", X_train.shape, X_test.shape, y_train.shape)
print("Saved: preprocessor_enriched.joblib, preprocessor_metadata_enriched.json, X_train.npy, X_test.npy, y_train.npy")
```

```python
# Full CV: evaluate enriched features (complete cell)
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import lightgbm as lgb
from lightgbm import early_stopping, log_evaluation
import gc, json, os

# Paths
DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csvs_dir = Path(DATA_DIR) / "csvs"
models_dir = Path(DATA_DIR) / "models"
os.makedirs(models_dir, exist_ok=True)

# Load preprocessed arrays (the enriched ones you just created)
X_train = np.load(csvs_dir / "X_train.npy")
X_test  = np.load(csvs_dir / "X_test.npy")
y       = np.load(csvs_dir / "y_train.npy")

print("Loaded shapes:", X_train.shape, X_test.shape, y.shape)

# Create stratify bins for regression
n_bins = 5
try:
    y_bins = pd.qcut(y, q=n_bins, labels=False, duplicates='drop')
except Exception:
    y_bins = pd.cut(y, bins=n_bins, labels=False)

# CV setup
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

oof_preds = np.zeros(len(y))
feature_importances = []
fold_metrics = []

# Conservative LightGBM params (you can tune later)
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'learning_rate': 0.02,
    'num_leaves': 16,
    'min_child_samples': 50,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 1,
    'lambda_l1': 1.0,
    'lambda_l2': 2.0,
    'verbosity': -1,
    'seed': 42,
    'n_jobs': -1
}

# CV loop
for fold, (train_idx, val_idx) in enumerate(skf.split(X_train, y_bins), start=1):
    print(f"\n===== Fold {fold} =====")
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y[train_idx], y[val_idx]

    lgb_train = lgb.Dataset(X_tr, label=y_tr)
    lgb_val   = lgb.Dataset(X_val, label=y_val, reference=lgb_train)

    model = lgb.train(
        params,
        lgb_train,
        num_boost_round=10000,
        valid_sets=[lgb_train, lgb_val],
        valid_names=['train','valid'],
        callbacks=[
            early_stopping(stopping_rounds=200, verbose=True),
            log_evaluation(period=200)
        ]
    )

    # Predictions and metrics
    val_pred = model.predict(X_val, num_iteration=model.best_iteration)
    oof_preds[val_idx] = val_pred

    rmse = np.sqrt(mean_squared_error(y_val, val_pred))
    pearson = pearsonr(y_val, val_pred)[0]
    print(f"Fold {fold} | RMSE: {rmse:.4f} | Pearson: {pearson:.4f} | best_iter: {model.best_iteration}")
    fold_metrics.append((fold, float(rmse), float(pearson), int(model.best_iteration)))

    # Feature importance (gain)
    try:
        fi = model.feature_importance(importance_type='gain')
    except Exception:
        fi = model.feature_importance()
    feature_importances.append(fi)

    # Save fold model
    model.save_model(str(models_dir / f"lgb_enriched_fold{fold}.txt"))
    del model, lgb_train, lgb_val
    gc.collect()

# Overall OOF metrics
oof_rmse = np.sqrt(mean_squared_error(y, oof_preds))
oof_pearson = pearsonr(y, oof_preds)[0]

print("\nPer-fold metrics (fold, rmse, pearson, best_iter):")
for fm in fold_metrics:
    print(fm)
print(f"\nOOF RMSE: {oof_rmse:.4f}")
print(f"OOF Pearson: {oof_pearson:.4f}")

# Save OOF preds
np.save(csvs_dir / "oof_preds_enriched.npy", oof_preds)
pd.DataFrame({"y_true": y, "y_pred": oof_preds}).to_csv(csvs_dir / "oof_preds_enriched.csv", index=False)

# Aggregate feature importance robustly (pad/truncate to match feature names)
# Attempt to read feature names from preprocessor metadata
feat_names = None
meta_candidates = [
    models_dir / "preprocessor_metadata_enriched.json",
    models_dir / "preprocessor_metadata_v2.json",
    models_dir / "preprocessor_metadata.json",
    models_dir / "preprocessor_info.json"
]
for mp in meta_candidates:
    if mp.exists():
        try:
            meta = json.load(open(mp))
            feat_names = meta.get("numeric_cols", None)
            if feat_names:
                break
        except Exception:
            feat_names = None

if feat_names is None:
    feat_names = [f"f{i}" for i in range(X_train.shape[1])]

# pad/truncate fi arrays to common length
fi_matrix = []
for fi in feature_importances:
    if fi.shape[0] == len(feat_names):
        fi_matrix.append(fi)
    elif fi.shape[0] > len(feat_names):
        fi_matrix.append(fi[:len(feat_names)])
    else:
        pad = np.zeros(len(feat_names), dtype=float)
        pad[:fi.shape[0]] = fi
        fi_matrix.append(pad)

fi_mean = np.mean(np.vstack(fi_matrix), axis=0)
fi_df = pd.DataFrame({"feature": feat_names, "gain": fi_mean})
fi_df = fi_df.sort_values("gain", ascending=False).reset_index(drop=True)
fi_df.to_csv(csvs_dir / "feature_importance_cv_enriched.csv", index=False)
print("Saved feature importance to:", csvs_dir / "feature_importance_cv_enriched.csv")
```

```python
# === Step: build fuzzy matches between features.csv and transcripts_whisper.csv ===
import difflib
import pandas as pd
from pathlib import Path

csv_dir = Path("/content/drive/MyDrive/SHL_dataset/dataset/csvs")
features_csv = csv_dir / "features.csv"
transcripts_csv = csv_dir / "transcripts_whisper.csv"
out_csv = csv_dir / "features_to_transcript_fuzzy_matches.csv"

feat = pd.read_csv(features_csv, low_memory=False)
trans = pd.read_csv(transcripts_csv, low_memory=False)

trans_fns = trans['filename'].astype(str).tolist()

rows = []
for fn in feat['filename'].astype(str):
    # exact match
    if fn in trans_fns:
        rows.append({"feat_fn": fn, "best_match": fn, "score": 1.0, "match_type": "exact"})
        continue

    # substring match
    substr = [t for t in trans_fns if t in fn or fn in t]
    if substr:
        best = max(substr, key=len)
        rows.append({"feat_fn": fn, "best_match": best, "score": 0.9, "match_type": "substring"})
        continue

    # fuzzy match via difflib
    bests = difflib.get_close_matches(fn, trans_fns, n=3, cutoff=0.1)
    if bests:
        best = bests[0]
        score = difflib.SequenceMatcher(None, fn, best).ratio()
        rows.append({"feat_fn": fn, "best_match": best, "score": score, "match_type": "fuzzy"})
    else:
        rows.append({"feat_fn": fn, "best_match": None, "score": 0.0, "match_type": "none"})

out = pd.DataFrame(rows)
out = out.sort_values("score", ascending=False)
out.to_csv(out_csv, index=False)

print(f"✅ Saved fuzzy matches to: {out_csv}")
print("Rows:", len(out))
print(out.head(25).to_string(index=False))
```

```python
# === STEP 2: LM + GEC feature extraction ===
# Installs (first run only):
# !pip install -q transformers==4.35.0 torch sentencepiece language_tool_python python-Levenshtein

import os, math
from pathlib import Path
import pandas as pd
import torch
from transformers import GPT2LMHeadModel, GPT2TokenizerFast
import language_tool_python, Levenshtein
from tqdm.auto import tqdm

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csv_dir  = Path(DATA_DIR) / "csvs"
in_csv   = csv_dir / "features_with_transcripts_autofill.csv"
out_csv  = csv_dir / "features_with_transcripts_lm_gec.csv"

df = pd.read_csv(in_csv, low_memory=False)
print("Loaded:", df.shape)

device = "cuda" if torch.cuda.is_available() else "cpu"
lm_name = "distilgpt2"       # faster; switch to "gpt2" for slightly better accuracy
print("device:", device, "| LM:", lm_name)

tokenizer = GPT2TokenizerFast.from_pretrained(lm_name)
model = GPT2LMHeadModel.from_pretrained(lm_name).to(device)
model.eval()

tool = language_tool_python.LanguageTool('en-US')

def gpt2_ppl(text, max_len=512):
    if not isinstance(text,str) or not text.strip(): return None,None
    toks = tokenizer.encode(text)
    losses = []
    for i in range(0, len(toks), max_len):
        chunk = toks[i:i+max_len]
        ids = torch.tensor([chunk], device=device)
        with torch.no_grad():
            out = model(ids, labels=ids)
        losses.append(out.loss.item())
    if not losses: return None,None
    mean_loss = sum(losses)/len(losses)
    ppl = math.exp(mean_loss) if mean_loss < 100 else float("inf")
    return mean_loss, ppl

losses, ppls, lt_edits, lt_corrs, lev_dists = [], [], [], [], []
for i, txt in enumerate(tqdm(df['transcript'].fillna("").tolist(), desc="Processing transcripts")):
    if not txt.strip():
        losses.append(None); ppls.append(None); lt_edits.append(None)
        lt_corrs.append(None); lev_dists.append(None)
        continue
    try:
        loss, ppl = gpt2_ppl(txt)
    except Exception as e:
        print(f"LM error at row {i}: {e}")
        loss, ppl = None, None
    losses.append(loss); ppls.append(ppl)

    try:
        matches = tool.check(txt)
        edit_ct = len(matches)
        corr = language_tool_python.utils.correct(txt, matches)
        lev = Levenshtein.distance(txt, corr)
    except Exception as e:
        edit_ct, corr, lev = None, None, None
        print(f"LT error at row {i}: {e}")
    lt_edits.append(edit_ct); lt_corrs.append(corr); lev_dists.append(lev)

df['lm_loss_distilgpt2'] = losses
df['ppl_distilgpt2']     = ppls
df['lt_edit_count']      = lt_edits
df['lt_correction']      = lt_corrs
df['levenshtein_chars']  = lev_dists

df['transcript_len_chars'] = df['transcript'].astype(str).map(len)
df['lt_edit_ratio_chars']  = df.apply(
    lambda r: (r['levenshtein_chars']/r['transcript_len_chars'])
              if pd.notnull(r['levenshtein_chars']) and r['transcript_len_chars']>0 else None,
    axis=1)

df.to_csv(out_csv, index=False)
print(f"\n✅ Saved LM+GEC enriched features to: {out_csv}")
print("Shape:", df.shape)

print("\nPreview of key columns:")
print(df[['filename','transcript_len_chars','ppl_distilgpt2',
          'lt_edit_count','lt_edit_ratio_chars']].head(10).to_string(index=False))
```

```python
# Re-run LightGBM CV with RMSE computed via np.sqrt(...) to avoid sklearn version issues
import os
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error
import lightgbm as lgb

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csv_dir  = Path(DATA_DIR) / "csvs"
in_csv   = csv_dir / "features_with_transcripts_lm_gec.csv"

df_full = pd.read_csv(in_csv, low_memory=False)

# determine label col
label_col = None
for cand in ['label','label_x','score','target','y','mos']:
    if cand in df_full.columns:
        label_col = cand
        break
if label_col is None:
    raise RuntimeError("Could not find label column automatically. Please tell me the label column name.")

# drop missing labels
df = df_full.loc[df_full[label_col].notna()].reset_index(drop=True)

# features selection
exclude_prefixes = ['filename','transcript','text','id','split']
excludes = [c for c in df.columns if any(p in c.lower() for p in exclude_prefixes)]
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
feature_cols = [c for c in num_cols if c != label_col and c not in excludes]
if 'has_transcript' not in feature_cols and 'has_transcript' in df.columns:
    feature_cols.append('has_transcript')

X = df[feature_cols].copy()
y = df[label_col].astype(float).copy()

# impute missing numeric with median
for c in X.columns:
    if X[c].isnull().any():
        med = X[c].median()
        X[c] = X[c].fillna(med)

# stratify bins
n_bins = 5
y_binned = pd.qcut(y, q=n_bins, labels=False, duplicates='drop')

# CV
SEED = 42
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
oof = np.zeros(len(X))
feature_importances = pd.DataFrame()
feature_importances['feature'] = feature_cols
rmse_scores = []

for fold, (tr_idx, val_idx) in enumerate(kf.split(X, y_binned)):
    X_tr, X_val = X.iloc[tr_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[tr_idx], y.iloc[val_idx]

    model = lgb.LGBMRegressor(
        n_estimators=2000,
        learning_rate=0.05,
        num_leaves=31,
        random_state=SEED,
        n_jobs=-1
    )
    model.fit(
        X_tr, y_tr,
        eval_set=[(X_val, y_val)],
        eval_metric='rmse',
        callbacks=[lgb.early_stopping(stopping_rounds=50), lgb.log_evaluation(period=100)]
    )

    val_pred = model.predict(X_val, num_iteration=model.best_iteration_)
    oof[val_idx] = val_pred
    rmse = np.sqrt(mean_squared_error(y_val, val_pred))
    rmse_scores.append(rmse)
    print(f"Fold {fold+1} RMSE: {rmse:.4f}")

    feature_importances[f'fold_{fold+1}'] = model.feature_importances_

# summary
print("\nCV RMSE scores:", rmse_scores)
print("Mean RMSE: {:.4f} | Std RMSE: {:.4f}".format(np.mean(rmse_scores), np.std(rmse_scores)))

feature_importances['importance_mean'] = feature_importances[[c for c in feature_importances.columns if c.startswith('fold_')]].mean(axis=1)
feat_imp_sorted = feature_importances.sort_values('importance_mean', ascending=False)
print("\nTop 15 feature importances:")
print(feat_imp_sorted[['feature','importance_mean']].head(15).to_string(index=False))

# save outputs
out_oof = csv_dir / "oof_preds_lgb_baseline_fixed_callbacks.csv"
out_imp = csv_dir / "lgb_feature_importances_baseline_fixed_callbacks.csv"
pd.DataFrame({'filename': df['filename'], 'y_true': y, 'y_pred': oof}).to_csv(out_oof, index=False)
feat_imp_sorted.to_csv(out_imp, index=False)
print("\nSaved OOF preds to:", out_oof)
print("Saved feature importances to:", out_imp)
```

```python
# === Final clean run (drop label_x and label_y) ===
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error
import lightgbm as lgb

DATA_DIR = "/content/drive/MyDrive/SHL_dataset/dataset"
csv_dir  = Path(DATA_DIR) / "csvs"
in_csv   = csv_dir / "features_with_transcripts_lm_gec.csv"

df_full = pd.read_csv(in_csv, low_memory=False)
label_col = 'label'
df = df_full.loc[df_full[label_col].notna()].reset_index(drop=True)

# drop leaked columns if present
drop_cols = [c for c in ['label_x','label_y'] if c in df.columns]
if drop_cols:
    df = df.drop(columns=drop_cols)
    print("Dropped leakage columns:", drop_cols)

# build features
exclude_prefixes = ['filename','transcript','text','id','split']
excludes = [c for c in df.columns if any(p in c.lower() for p in exclude_prefixes)]
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
feature_cols = [c for c in num_cols if c != label_col and c not in excludes]
if 'has_transcript' not in feature_cols and 'has_transcript' in df.columns:
    feature_cols.append('has_transcript')

X = df[feature_cols].copy()
y = df[label_col].astype(float).copy()

# impute missing numeric with median
for c in X.columns:
    if X[c].isnull().any():
        X[c] = X[c].fillna(X[c].median())

# stratify bins
y_binned = pd.qcut(y, q=5, labels=False, duplicates='drop')

# cross-validation
SEED = 42
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
oof = np.zeros(len(X))
feature_importances = pd.DataFrame({'feature': feature_cols})
rmse_scores = []

for fold, (tr_idx, val_idx) in enumerate(kf.split(X, y_binned)):
    X_tr, X_val = X.iloc[tr_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[tr_idx], y.iloc[val_idx]

    model = lgb.LGBMRegressor(
        n_estimators=2000,
        learning_rate=0.05,
        num_leaves=31,
        random_state=SEED,
        n_jobs=-1
    )
    model.fit(
        X_tr, y_tr,
        eval_set=[(X_val, y_val)],
        eval_metric='rmse',
        callbacks=[lgb.early_stopping(stopping_rounds=50), lgb.log_evaluation(period=100)]
    )

    val_pred = model.predict(X_val, num_iteration=model.best_iteration_)
    oof[val_idx] = val_pred
    rmse = np.sqrt(mean_squared_error(y_val, val_pred))
    rmse_scores.append(rmse)
    print(f"Fold {fold+1} RMSE: {rmse:.4f}")

    feature_importances[f'fold_{fold+1}'] = model.feature_importances_

print("\nCV RMSE scores:", rmse_scores)
print("Mean RMSE: {:.4f} | Std RMSE: {:.4f}".format(np.mean(rmse_scores), np.std(rmse_scores)))

feature_importances['importance_mean'] = feature_importances[[c for c in feature_importances.columns if c.startswith('fold_')]].mean(axis=1)
feat_imp_sorted = feature_importances.sort_values('importance_mean', ascending=False)
print("\nTop 15 true feature importances (leakage-free):")
print(feat_imp_sorted[['feature','importance_mean']].head(15).to_string(index=False))
```
